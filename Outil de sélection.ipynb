Sélection d'une équipe type sans contrainte spécifique

```python
#Import des bibliothèques nécessaires

import pandas as pd

#Import des données
df = pd.read_csv(r"C:\Users\Dylan\Desktop\Portfolio\MPG\MPG_NOT_ENCODED.csv")

#Modification type des colonnes
df['Valeur_Joueur'] = df['Valeur_Joueur'].astype(int)
df['Note'] = df['Note'].astype(int)

#Budget maximum disponible
budget_maximum = 500
prix_maximum_par_joueur = 75

#Tri par la colonne Note
df = df.sort_values(by='Note', ascending=False)

#Création des listes et des dictionnaires
meilleurs_joueurs = []
cout_total = 0
equipe_selectionnee = {'G': [], 'DC': [], 'DL': [], 'MD': [], 'MO': [], 'A': []}
postes_requis = {'G': 4, 'DC': 3, 'DL': 3, 'MD': 3, 'MO': 3, 'A': 2}

#Itération des joueurs
for index, joueur in df.iterrows():
    #Vérification si l'ajout de ce joueur ne dépasse pas le budget maximum et si le prix du joueur est inférieur 
    #ou égal au prix maximum par joueur
    if cout_total + joueur['Valeur_Joueur'] <= budget_maximum and joueur['Valeur_Joueur'] <= prix_maximum_par_joueur:
        #Vérification si le poste du joueur est requis
        poste = joueur['Poste']
        if poste in postes_requis and postes_requis[poste] > 0:
            meilleurs_joueurs.append(joueur)
            cout_total += joueur['Valeur_Joueur']
            equipe_selectionnee[poste].append(joueur['Joueur'])
            postes_requis[poste] -= 1

    #Si nous avons sélectionné 18 joueurs, nous pouvons sortir de la boucle
    if len(meilleurs_joueurs) == 18:
        break
        
#Afficher le nom de l'équipe    
print(f"Equipe Neutre")

#Afficher  la liste des 18 meilleurs joueurs
for i, joueur in enumerate(meilleurs_joueurs, start=1):
    print(f"Joueur {i}: {joueur['Joueur']} - Poste: {joueur['Poste']}, Note: {joueur['Note']}, Prix: {joueur['Valeur_Joueur']}")

#Afficher  le coût total
print(f"Coût total des 18 meilleurs joueurs: {cout_total}")

#Afficher les joueurs sélectionnés par poste
for poste, joueurs in equipe_selectionnee.items():
    print(f"Poste {poste}: {', '.join(joueurs)}")
```

Sélection d'une équipe avec un budget poussé à 100 maximum pour récupérer les meilleurs attaquants

```python
#Import des données
df = pd.read_csv(r"C:\Users\Dylan\Desktop\Portfolio\MPG\MPG_NOT_ENCODED.csv")

#Modification type des colonnes
df['Valeur_Joueur'] = df['Valeur_Joueur'].astype(int)
df['Note'] = df['Note'].astype(int)

#Budget maximum disponible
budget_maximum = 500
prix_maximum_par_joueur = 200
prix_maximum_par_poste = {'A': 100, 'DC': 50, 'DL': 50, 'MD': 50, 'MO': 50, 'G': 20}  # Prix maximum par poste

#Tri par la colonne Note
df = df.sort_values(by='Note', ascending=False)

#Création des listes et des dictionnaires
meilleurs_joueurs = []
cout_total = 0
equipe_selectionnee = {'A': [], 'DC': [], 'DL': [], 'MD': [], 'MO': [], 'G': []}
postes_requis = {'G': 4, 'DC': 3, 'DL': 3, 'MD': 3, 'MO': 3, 'A': 2}

#Itération des joueurs
for index, joueur in df.iterrows():
    #Vérification si l'ajout de ce joueur ne dépasse pas le budget maximum,
    #si le prix du joueur est inférieur ou égal au prix maximum par joueur,
    #et si le prix du joueur est inférieur ou égal au prix maximum par poste pour ce poste
    if (
        cout_total + joueur['Valeur_Joueur'] <= budget_maximum
        and joueur['Valeur_Joueur'] <= prix_maximum_par_joueur
        and joueur['Valeur_Joueur'] <= prix_maximum_par_poste[joueur['Poste']]
    ):
        #Vérification si le poste du joueur est requis
        poste = joueur['Poste']
        if poste in postes_requis and postes_requis[poste] > 0:
            meilleurs_joueurs.append(joueur)
            cout_total += joueur['Valeur_Joueur']
            equipe_selectionnee[poste].append(joueur['Joueur'])
            postes_requis[poste] -= 1

    #Si nous avons sélectionné 18 joueurs, nous pouvons sortir de la boucle
    if len(meilleurs_joueurs) == 18:
        break

#Afficher le nom de l'équipe    
print(f"Equipe Attaque")

#Affichez la liste des 18 meilleurs joueurs
for i, joueur in enumerate(meilleurs_joueurs, start=1):
    print(f"Joueur {i}: {joueur['Joueur']} - Poste: {joueur['Poste']}, Note: {joueur['Note']}, Prix: {joueur['Valeur_Joueur']}")

#Affichez le coût total
print(f"Coût total des 18 meilleurs joueurs: {cout_total}")

#Affichez les joueurs sélectionnés par poste
for poste, joueurs in equipe_selectionnee.items():
    print(f"Poste {poste}: {', '.join(joueurs)}")
```
